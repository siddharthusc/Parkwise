```python
import pandas as pd
from sodapy import Socrata
```

```python
%%time
# Example authenticated client (needed for non-public datasets):
# client = Socrata(data.lacity.org,
#                  MyAppToken,
#                  username="user@example.com",
#                  password="AFakePassword")

client = Socrata("data.lacity.org",
                 "8JwvqNkTSYSyJhCJckySNhwx4")
socrata_raw = client.get("wjz9-h9np", where="issue_date BETWEEN '2020-01-01' AND '2023-03-29'", limit=16000000)
# Convert to pandas DataFrame
df = pd.DataFrame.from_records(socrata_raw)
```

Since the full dataset is too large, we fetch citation records that starts at 2020-01-01.
